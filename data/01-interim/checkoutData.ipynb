The `CRCP_Benthic_Cover_Florida_7018_0ee5_9488.csv` dataset contains 13,602 benthic survey records from Florida coral reefs collected between 2016 and 2022. Each row represents an observation at a specific station and date, with geographic coordinates (`latitude`, `longitude`), regional and zoning information (`REGION`, `SUB_REGION_NAME`, `ZONE_NAME`, `MPA_NAME`), and details about the survey such as depth, transect length, and reef complexity (`RUGOSITY_CD`, `WTD_RUG`). Benthic cover is recorded by category (`COVER_CAT_NAME`) along with percent cover of hard bottom, soft bottom, and rubble. Several columns contain missing values or are stored as text rather than numeric values, and the first row of the file contains unit labels rather than actual data. The dataset is suitable for spatial and temporal analysis of coral reef benthic composition, habitat type distribution, and substrate characteristics across Florida reef sites.

### What the datasets contain
Across the three regions, the datasets provide:
* ~50,000+ benthic survey records total
* Time span:
  * Florida: 2016–2022
  * Puerto Rico: 2014–2023
  * USVI: 2013–2023

* Spatial coverage:
  * Florida Keys
  * Puerto Rico reef systems
  * U.S. Virgin Islands reef systems

* Each row represents:
  * A specific station
  * On a specific date
  * With a specific benthic cover category
  * And associated percent substrate cover


### Key Things for Our Question
1. Percent Live Coral Cover
   * `COVER_CAT_NAME` identifies species or benthic groups
   * `HARDBOTTOM_P`, `SOFTBOTTOM_P`, and `RUBBLE_P` contain percent cover values
    > By filtering for live coral species and summing their percent cover by station and year, we can calculate total live coral cover and track changes over time

2. Tempurature Information
   * `YEAR`, `MONTH`, and `Date_UTC` allow us to build time series of coral cover
    > These variables enable analysis of decline and recovery patterns following stress events
3. Algal Cover
   * Algal groups are also listed under COVER_CAT_NAME (e.g., macroalgae, turf algae, cyanobacteria)
  > Summing these categories allows comparison of coral and algal trends to detect potential phase shifts
4. Reef Complexity
   * WTD_RUG and MEAN_RUG measure structural complexity
  > These variables can help assess whether more structurally complex reefs recover differently than flatter reefs
