# Research Question

Since the 2016 mass coral bleaching event, what distinct recovery trajectories in percent live coral cover are observed across U.S. Caribbean coral reef sites during 2016-2020 (a four-year recovery window in two intervals of two years each), and how are cumulative thermal stress and mean algal cover associated with these different recovery pathways?

## Background and Prior Work

Coral reefs are highly sensitive marine ecosystems that provide important ecological, economic, and coastal protection benefits. In recent years, coral reef health has declined worldwide due to rising sea surface temperatures, more frequent marine heatwaves, and local stressors such as algal overgrowth. Thermal stress, often measured using Degree Heating Weeks (DHW), is strongly linked to coral bleaching events, which can lead to partial or complete loss of live coral cover. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) While many studies document coral decline following thermal stress, less is known about how coral reefs recover over time and weather recovery follows consistent patterns across different sites.  <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Being students of University of California, San Diego, part of our academic environment is ingrained with the university’s strengths in marine biology, oceanography, and climate science through programs such as the Scripps Institution of Oceanography. UC San Diego researchers have made major contributions to coral reef monitoring and the study of climate-driven marine stressors, which does add to the motivation for this project. Publicly available datasets from organizations like NOAA and the National Coral Reef Monitoring Program align well with UC San Diego’s emphasis on data-driven marine science and provide a strong foundation for this analysis.

Previous research has shown that coral recovery trajectories can vary widely depending on environmental conditions and local ecological dynamics. Studies using NOAA Coral Reef Watch data have found that higher DHW values are associated with more severe bleaching and increased coral mortality, often resulting in slow or incomplete recovery. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) However, more recent monitoring efforts, such as the National Coral Reef Monitoring Program (NCRMP), provide long-term, site-level data on percent live coral and algal cover across U.S. reef systems. <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

While many studies using these data focus on overall trends in reef health, fewer explicitly examine differences in recovery pathways over time. Our project builds on this prior work by identifying coral recovery trajectories and analyzing how thermal stress and algal cover, both individually and together, are associated with differences in recovery outcomes across U.S. coral reef sites.

### Dataset #2
## Puerto Rico Coral Reef Benthic Cover (CRCP Monitoring Data, 2014–2023)
---

The Puerto Rico Benthic Cover dataset includes 10,656 reef survey observations collected between 2014 and 2023 through NOAA’s Coral Reef Conservation Program. Each row represents one benthic category (like live coral or algae) measured at a specific reef site on a particular survey date. The dataset includes the site ID (`PRIMARY_SAMPLE_UNIT`), latitude and longitude (in decimal degrees), survey year (and month/day when available), region, the organism category (`COVER_CAT_NAME`), and the percent of hardbottom covered (`HARDBOTTOM_P`).

The main variable we care about is percent hardbottom cover, which is measured in percent (%). This tells us how much of the reef surface is covered by a certain organism type. For example, 30% live coral cover means that about one-third of the surveyed reef area was covered by living coral. In many reef systems, coral cover above about 30% is considered relatively healthy, while values below 10% often suggest significant degradation or limited recovery after stress events.

Since our research question looks at recovery patterns in live coral cover and how those patterns relate to algal cover, we use COVER_CAT_NAME to separate coral from algae and HARDBOTTOM_P to measure how dominant each group is. By combining location (site and region) with time (year), we can track how coral cover changes over time at different sites and see whether some reefs recover differently than others. This helps us explore whether higher algal cover is associated with weaker coral recovery after stress events.

**Variables We're Focusing On**

For this analysis, we’re mainly looking at location and habitat structure variables from the benthic dataset.
* Site ID (`PRIMARY_SAMPLE_UNIT`)
This is the unique identifier for each survey site. It allows us to group observations by location and track changes over time at the same site
* Latitude (degrees north) and Longitude (degrees east)
These tell us where each survey site is located. We’ll use them to look at spatial patterns and see whether habitat characteristics cluster in certain areas.
* Date (`YEAR`, `MONTH`, `DAY`)
These variables describe when the observation was recorded. YEAR allows us to evaluate long-term trends, while month/day could help identify seasonal patterns.
* Region 
This categorizes sites into broader geographic areas within Puerto Rico. It allows us to compare coral and algae cover across different parts of the island.
* `COVER_CAT_NAME`
This identifies the organism category being measured (e.g., coral, algae). This is central to our research question because we are specifically comparing coral vs. algae cover.
* Percent Cover (`HARDBOTTOM_P`)
This is the key outcome variable. It represents the percentage of hardbottom substrate covered by the organism listed in `COVER_CAT_NAME`.
It is measured as a percent (%), ranging theoretically from 0 to 100.
> * Values near 0% indicate little to no coverage.
> * Values near 100% indicate nearly complete coverage of the hardbottom area by that organism type.
> * Values outside 0–100% would be biologically impossible and indicate data issues.

By combining site location (latitude and longitude), region, time (year, month, day), organism category (coral vs. algae), and percent hardbottom cover, we can analyze how benthic community composition varies across Puerto Rico and over time. These variables allow us to examine both spatial differences between regions and temporal trends within sites. Rather than simply describing individual survey locations, this approach helps us identify broader patterns in coral and algae cover and assess how benthic communities may be shifting across space and time.


**Concerns and Limitations**

One major concern is sampling bias. Survey sites are not randomly distributed across all reef habitat — they are likely chosen based on accessibility, monitoring priorities, or ecological importance. This means results may not represent all reef areas equally.

Another limitation is temporal imbalance. Some sites may have been surveyed more frequently than others, and some years may have more complete coverage than others. This could affect trend analysis.

In addition, several numeric variables (like latitude, depth, and percent cover values) are stored as text and need to be converted before analysis. The first row of the dataset also contains unit labels instead of real data and must be removed during cleaning.

Finally, the dataset only covers about 10 years (2014–2023), not the full 20-year period in our research question. While it still captures recent bleaching and stress events, longer-term trends would require additional historical data.

```python
CORAL_SPECIES = {
    # Individual coral species
    "Eusmilia fastigiata", "Meandrina meandrites", "Orbicella faveolata",
    "Siderastrea siderea", "Stephanocoenia intersepta", "Millepora spp",
    "Dichocoenia stokesii", "Scolymia spp", "Montastraea cavernosa",
    "Colpophyllia natans", "Siderastrea radians", "Pseudodiploria strigosa",
    "Porites astreoides", "Diploria labyrinthiformis", "Orbicella annularis",
    "Agaricia agaricites", "Mycetophyllia lamarckiana", "Oculina robusta",
    "Mussa angulosa", "Porites porites", "Mycetophyllia aliciae",
    "Scolymia cubensis", "Acropora cervicornis", "Porites divaricata",
    "Solenastrea bournoni", "Siderastrea spp", "Madracis decactis",
    "Agaricia spp", "Porites spp", "Solenastrea spp", "Orbicella franksi",
    "Agaricia lamarcki", "Other coral", "Dendrogyra cylindrus",
    "Agaricia humilis", "Agaricia fragilis", "Favia fragum", "Meandrina spp",
    "Diploria spp.", "Colpophyllia spp", "Mycetophyllia ferox",
    "Helioceris cucullata", "Oculina diffusa", "Pseudodiploria clivosa",
    "Manicina areolata", "Madracis auretenra", "Madracis spp",
    "Porites furcata", "Madracis formosa", "Madracis carmabi",
    "Isophyllia sinuosa", "Acropora palmata",
}

ALGAE_SPECIES = {
    "Turf Algae with Sediment", "Turf Algae Free of Sediment",
    "Dictyota spp", "MacroOtherFleshy", "Lobophora spp",
    "Cyanophyta spp", "Peysonnellia", "Halimeda spp",
    "MacroOtherCalcareous", "Ramicrusta spp", "Rhodophyta cru. spp",
    "Magnoliophyta spp",
}
```

```python
## Filter to Relevant Categories (Coral and Algae)

# Function to classify categories
def classify_category(cat_name):
    if pd.isna(cat_name):
        return None
    cat = cat_name.strip()
    # Pass 1: exact match
    if cat in CORAL_SPECIES:
        return "Coral"
    if cat in ALGAE_SPECIES:
        return "Algae"
    # Pass 2: substring on uppercase
    upper = cat.upper()
    coral_keywords = ["CORAL", "STONY CORAL", "SCL CORAL", "LIVE CORAL"]
    algae_keywords = ["ALGAE", "MACROALG", "TURF", "CYANO", "HALIMEDA",
                      "DICTYOTA", "LOBOPHORA", "RAMICRUSTA", "RHODOPHYTA",
                      "PEYSSONNELIA", "PEYSONNELLIA"]
    if any(kw in upper for kw in coral_keywords):
        return "Coral"
    if any(kw in upper for kw in algae_keywords):
        return "Algae"
    return None

# Apply classification
for df, name in [(florida_benthic, "Florida"),
                 (puerto_rico_benthic, "Puerto Rico"),
                 (usvi_benthic, "USVI")]:
    df["group"] = df["COVER_CAT_NAME"].apply(classify_category)
    coral_n = (df["group"] == "Coral").sum()
    algae_n = (df["group"] == "Algae").sum()
    total   = len(df)
    print(f"{name}: {coral_n} coral rows, {algae_n} algae rows "
          f"({100*(coral_n+algae_n)/total:.1f}% of total classified)")
```

We now have only the rows that are useful for analyzing coral vs. algal percent cover.

```python
## Aggregate by Site and Year

def aggregate_site_year(df):
    df = df[df['group'].isin(['Coral', 'Algae'])].copy()

    site_agg = df.groupby(
        ['PRIMARY_SAMPLE_UNIT', 'YEAR', 'group']
    ).agg(
        HARDBOTTOM_P=('HARDBOTTOM_P', 'mean'),
        latitude=('latitude', 'first'),
        longitude=('longitude', 'first'),
        REGION=('REGION', 'first'),
    ).reset_index()

    pivoted = site_agg.pivot_table(
        index=['PRIMARY_SAMPLE_UNIT', 'YEAR', 'latitude', 'longitude', 'REGION'],
        columns='group',
        values='HARDBOTTOM_P',
        fill_value=0
    ).reset_index()

    pivoted.columns.name = None
    pivoted = pivoted.rename(columns={
        'Coral': 'total_coral_cover',
        'Algae': 'total_algae_cover'
    })
    return pivoted
    
florida_site_year     = aggregate_site_year(florida_benthic)
puerto_rico_site_year = aggregate_site_year(puerto_rico_benthic)
usvi_site_year        = aggregate_site_year(usvi_benthic)

site_year_cover = pd.concat(
    [florida_site_year, puerto_rico_site_year, usvi_site_year],
    ignore_index=True,
)

print(f"\nFinal dataset shape: {site_year_cover.shape}")
print(f"Years covered: {sorted(site_year_cover['YEAR'].dropna().astype(int).unique())}")
print(f"Regions: {site_year_cover['REGION'].unique()}")
print(f"\nCoral cover stats:\n{site_year_cover['total_coral_cover'].describe().round(2)}")
print(f"\nAlgae cover stats:\n{site_year_cover['total_algae_cover'].describe().round(2)}")
site_year_cover.head()
```

This table now has one row per site-year, showing the total coral and algae cover, along with latitude, longitude, and region. This is exactly what the other team members need for their EDA.

```python
benthic_all = benthic_all.dropna(subset=['YEAR'])
benthic_all['YEAR'] = benthic_all['YEAR'].astype(int)

years_to_use = [2016, 2018, 2020]
benthic_all = benthic_all[benthic_all['YEAR'].isin(years_to_use)]
print(f" Benthic data : {len(benthic_all)} rows")
```

Aggregate benthic data to site-year 

```python
site_year_benthic = benthic_all.groupby(['PRIMARY_SAMPLE_UNIT', 'YEAR', 'latitude', 'longitude', 'REGION']).agg({
    'HARDBOTTOM_P': 'sum'
}).reset_index()

site_year_benthic.columns = ['PRIMARY_SAMPLE_UNIT', 'YEAR', 'latitude', 'longitude', 'REGION', 'total_cover']
print(f"Aggregated to {len(site_year_benthic)} site-year observations")
print(site_year_benthic.head() )
```

```python
thermal_data['year'] = thermal_data['year'].astype(int)
thermal_data = thermal_data[thermal_data['year'].isin(years_to_use)]
print(f"Thermal data loaded: {len(thermal_data)} rows")

# Converting coordinates to numeric and clean
print("\nSTEP 4: Cleaning coordinates...")
site_year_benthic['latitude'] = pd.to_numeric(site_year_benthic['latitude'], errors='coerce')
site_year_benthic['longitude'] = pd.to_numeric(site_year_benthic['longitude'], errors='coerce')
thermal_data['latitude'] = pd.to_numeric(thermal_data['latitude'], errors='coerce')
thermal_data['longitude'] = pd.to_numeric(thermal_data['longitude'], errors='coerce')

site_year_benthic = site_year_benthic.dropna(subset=['latitude', 'longitude'])
thermal_data = thermal_data.dropna(subset=['latitude', 'longitude'])
print(f" After cleaning: {len(site_year_benthic)} benthic, {len(thermal_data)} thermal")

# Merging by year with distance threshold
print("\nSTEP 5: Merging benthic + thermal data...")
merged_results = []

for year in sorted(years_to_use):
    benthic_year = site_year_benthic[site_year_benthic['YEAR'] == year].copy()
    thermal_year = thermal_data[thermal_data['year'] == year].copy()
    
    print(f"\n  {int(year)}:")
    print(f"    Benthic sites: {len(benthic_year)}")
    print(f"    Thermal grid points: {len(thermal_year)}")
    
    # Building BallTree
    thermal_coords_rad = np.radians(thermal_year[['latitude', 'longitude']].values)
    tree = BallTree(thermal_coords_rad, metric='haversine')
    
    # Nearest neighbor
    benthic_coords_rad = np.radians(benthic_year[['latitude', 'longitude']].values)
    distances, indices = tree.query(benthic_coords_rad, k=1)
    
    # Converting to km
    distances_km = distances.flatten() * 6371
    
    # Distance threshold = 5 km 
    distance_threshold = 5.0
    valid_matches = distances_km <= distance_threshold
    
    print(f"    Matches within {distance_threshold}km: {valid_matches.sum()} / {len(benthic_year)} ({valid_matches.sum()/len(benthic_year)*100:.1f}%)")
    print(f"    Distance stats: mean={distances_km.mean():.2f}km, max={distances_km.max():.2f}km")
    
    # Getting valid matches
    valid_indices = np.where(valid_matches)[0]
    valid_thermal_indices = indices.flatten()[valid_matches]
    
    benthic_year_valid = benthic_year.iloc[valid_indices].copy()
    thermal_year_subset = thermal_year.iloc[valid_thermal_indices].copy()
    
    benthic_year_valid['annual_dhw'] = thermal_year_subset['annual_dhw'].values
    benthic_year_valid['distance_to_thermal_km'] = distances_km[valid_matches]
    
    dhw_stats = benthic_year_valid['annual_dhw']
    print(f"    DHW stats: mean={dhw_stats.mean():.2f}°C-weeks, max={dhw_stats.max():.2f}°C-weeks")
    
    merged_results.append(benthic_year_valid)
```

# Coral Cover vs Thermal Stress 

```python
print("\nCoral Cover Statistics:")
print(f"  Mean coral: {merged_final['total_cover'].mean():.2f}%")
print(f"  Median coral: {merged_final['total_cover'].median():.2f}%")
print(f"  Range: {merged_final['total_cover'].min():.2f} - {merged_final['total_cover'].max():.2f}%")

# By stress level
low_stress = merged_final[merged_final['annual_dhw'] < 4]
high_stress = merged_final[merged_final['annual_dhw'] >= 4]

print(f"\nCoral by stress level:")
print(f"  Low stress (DHW < 4): n={len(low_stress)}, mean coral={low_stress['total_cover'].mean():.2f}%")
print(f"  High stress (DHW ≥ 4): n={len(high_stress)}, mean coral={high_stress['total_cover'].mean():.2f}%")
print(f"  Difference: {high_stress['total_cover'].mean() - low_stress['total_cover'].mean():.2f}%")


correlation = merged_final['annual_dhw'].corr(merged_final['total_cover'])
print(f"\nPearson correlation (DHW vs coral): r = {correlation:.3f}")


fig, axes = plt.subplots(1, 2, figsize=(14, 5))


ax1 = axes[0]
stress_categories = ['Low stress\n(DHW < 4)', 'High stress\n(DHW ≥ 4)']
coral_by_stress = [low_stress['total_cover'].values, high_stress['total_cover'].values]

bp = ax1.boxplot(coral_by_stress, labels=stress_categories, patch_artist=True, widths=0.6)
for patch, color in zip(bp['boxes'], ['lightgreen', 'lightcoral']):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax1.set_ylabel('Total Coral Cover (%)', fontsize=12)
ax1.set_title('Coral Cover by Thermal Stress Level\n(Bleaching threshold: DHW ≥ 4°C-weeks)', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')


ax2 = axes[1]
colors_by_year = {2016.0: '#d62728', 2018.0: '#ff7f0e', 2020.0: '#2ca02c'}
for year in sorted(merged_final['YEAR'].unique()):
    year_data = merged_final[merged_final['YEAR'] == year]
    ax2.scatter(year_data['annual_dhw'], year_data['total_cover'], 
               label=f'{int(year)} (n={len(year_data)})', 
               s=80, alpha=0.6, color=colors_by_year[year], edgecolors='black', linewidth=0.5)


z = np.polyfit(merged_final['annual_dhw'], merged_final['total_cover'], 1)
p = np.poly1d(z)
x_trend = np.linspace(merged_final['annual_dhw'].min(), merged_final['annual_dhw'].max(), 100)
ax2.plot(x_trend, p(x_trend), "k--", linewidth=2, label=f'Trend (r={correlation:.2f})')

ax2.axvline(4, color='red', linestyle='--', linewidth=1.5, alpha=0.5, label='Bleach threshold')
ax2.set_xlabel('Annual Maximum DHW (°C-weeks)', fontsize=12)
ax2.set_ylabel('Total Coral Cover (%)', fontsize=12)
ax2.set_title('Coral Cover vs Thermal Stress\n(Relationship test)', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('02_coral_vs_dhw.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from scipy.stats import pearsonr, ttest_ind
corr_coef, p_value = pearsonr(merged_final['annual_dhw'], merged_final['total_cover'])
print(f"\nStatistical significance:")
print(f"  Pearson r = {corr_coef:.3f}")
print(f"  p-value = {p_value:.4f}")
if p_value < 0.05:
    print(f" Significant relationship (p < 0.05)")
else:
    print(f" Not significant (p ≥ 0.05)")

# T-test between groups
t_stat, t_pval = ttest_ind(low_stress['total_cover'], high_stress['total_cover'])
print(f"\nT-test (Low vs High stress coral cover):")
print(f"  t-statistic = {t_stat:.3f}")
print(f"  p-value = {t_pval:.4f}")
if t_pval < 0.05:
    print(f" Significant difference between groups (p < 0.05)")
else:
    print(f" No significant difference (p ≥ 0.05)")
```

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(merged_final['annual_dhw'], 
                                                           merged_final['total_cover'])

print(f"Regression equation: Coral Cover = {intercept:.2f} + {slope:.4f} × DHW")
print(f"Slope: {slope:.4f}")
```

Correlation Analysis:

- p-value: 3.05 × 10⁻⁹ (highly significant)
This helps us conclude that as thermal stress increases, coral cover decreases. 

While the correlation is weak, the relationship is highly significant statistically. This means the pattern is real and not due to chance—even though other factors explain most of the variation.

The negative correlation indicates that reef sites experiencing higher thermal stress tend to have lower coral cover. A 1°C-week increase in DHW is associated with approximately 0.07% decrease in coral cover (based on the regression slope).

```python
for year in sorted(merged_final['YEAR'].unique()):
    year_data = merged_final[merged_final['YEAR'] == year]['total_cover']
    print(f"  {int(year)}: mean={year_data.mean():.2f}%, median={year_data.median():.2f}%, n={len(year_data)}")

# Calculating recovery rate
coral_2016 = merged_final[merged_final['YEAR'] == 2016]['total_cover'].mean()
coral_2018 = merged_final[merged_final['YEAR'] == 2018]['total_cover'].mean()
coral_2020 = merged_final[merged_final['YEAR'] == 2020]['total_cover'].mean()

print(f" Recovery trajectory:")
print(f"  2016 to 2018 change: {coral_2018 - coral_2016:.2f}% ({(coral_2018-coral_2016)/coral_2016*100:.1f}% per year)")
print(f"  2018 to 2020 change: {coral_2020 - coral_2018:.2f}% ({(coral_2020-coral_2018)/2/coral_2018*100:.1f}% per year)")
print(f"  2016 to 2020 change: {coral_2020 - coral_2016:.2f}% ({(coral_2020-coral_2016)/4/coral_2016*100:.1f}% per year)")


fig, ax = plt.subplots(figsize=(10, 6))

data_by_year = [merged_final[merged_final['YEAR'] == year]['total_cover'].values 
                 for year in sorted(merged_final['YEAR'].unique())]

bp = ax.boxplot(data_by_year, 
                labels=[f'{int(y)}' for y in sorted(merged_final['YEAR'].unique())],
                patch_artist=True,
                widths=0.6)

colors = ['#d62728', '#ff7f0e', '#2ca02c']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_xlabel('Year', fontsize=12, fontweight='bold')
ax.set_ylabel('Total Coral Cover (%)', fontsize=12, fontweight='bold')
ax.set_title('Coral Cover Recovery Over Time\n(2016 bleaching year → 2020 recovery)', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('03_coral_by_year.png', dpi=300, bbox_inches='tight')
plt.show()
```

All three years show wide ranges in coral cover (0% to 400%), indicating substantial heterogeneity across reef sites. Some reefs maintained healthy coral coverage throughout the period while some other reefs had severely depleted coral cover. 

The median shows remarkable stability, suggesting no major shifts in typical reef condition. 
Many sites with near-zero coral cover appear as circles below the whiskers. These outliers are consistent across all three years which suggests that some reefs are chronically degraded and that this degradation is probablu unrelated to the 2016 bleaching event. 

Looking at the recovery trajectory, there is a lack of a clear upward recovery trend, which goes against our hypothesis of seeing a clear, albeit slow recovery pattern post bleaching. Recovery trajectories exist but do not have a clear pattern to them and are heterogeneous. 

Thermal stress varies substantially (0.90 -> 0.49 -> 0.00°C-weeks from 2016-2020), yet coral cover shows minimal change. This decoupling between thermal stress and recovery suggests that thermal stress alone does not determine recovery trajectory in these reefs.
